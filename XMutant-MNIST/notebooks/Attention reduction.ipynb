```python


def attention_reduction(subpaths):
    #diffs = np.array([])
    diffs_rel = np.array([])
    
    for subpath in subpaths:
        csv_path = os.path.join(subpath, "individual_logs/0.csv")
        if os.path.exists(csv_path):
            df = pd.read_csv(csv_path)
            if len(df) > 2:
                #print(df["avg_intensity_before"][1:-1].to_numpy())
                #print(df["avg_intensity_after"][2:].to_numpy())
                #print(df["avg_intensity_before"][1:-1].to_numpy() - df["avg_intensity_after"][2:].to_numpy())
                #diffs = np.concatenate((diffs, df["avg_intensity_before"][1:-1].to_numpy() - df["avg_intensity_after"][2:].to_numpy()))

                diffs_rel = np.concatenate((diffs_rel, df["rel_intensity_before"][1:-1].to_numpy() - df["rel_intensity_after"][2:].to_numpy()))
        else:
            print(csv_path, "does not exist")
    #print("total reduction")
    ##print(diffs)
    #print(diffs.mean())
    #print((diffs>0).sum()/len(diffs))

    print("relative reduction")
    print(diffs_rel)
    print(diffs_rel.mean())
    print((diffs_rel>0).sum()/len(diffs_rel))
```
