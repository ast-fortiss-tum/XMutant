```python
def merge_time(dict1, dict2):
    merged_time = dict1.copy()

    for key, value in dict2.items():
        if key in merged_time:
            merged_time[key] += value
    return merged_time
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
methods = ['FGSM', 'DLFuzz', 'DeepXplore', 'XMutant']
misclassification_rates = [43.8, 60.1, 44/2000, 74.15]  # %
time_per_seed = [0.37, 0.65, (56*60+21)/44, 0.3277]  # seconds

x = np.arange(len(methods))  # x-axis positions
width = 0.35  # bar width

# Create figure and axes
fig, ax1 = plt.subplots(figsize=(10, 3))

# Plot Misclassification Rate (left axis)
bars1 = ax1.bar(x - width/2, misclassification_rates, width,
                color='#1f77b4', edgecolor='black', label='Misclassification Rate (%)')
ax1.set_xlabel('Methods', fontsize=12)
ax1.set_ylabel('Misclassification Rate (%)', color='#1f77b4', fontsize=12)
ax1.tick_params(axis='y', labelcolor='#1f77b4')
ax1.set_ylim(0, 100)
ax1.set_xticks(x)
ax1.set_xticklabels(methods, fontsize=12)

# Annotate misclassification rates
for bar in bars1:
    height = bar.get_height()
    ax1.annotate(f'{height:.2f}%',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Create second y-axis for Time per Seed
ax2 = ax1.twinx()
bars2 = ax2.bar(x + width/2, time_per_seed, width,
                color='#ff7f0e', edgecolor='black', label='Time per Seed (s)')
ax2.set_ylabel('Time per Seed (seconds)', color='#ff7f0e', fontsize=12)
ax2.tick_params(axis='y', labelcolor='#ff7f0e')
ax2.set_ylim(0, 7.5)  # Adjusted to show DeepXplore's 6.8s clearly

# Annotate time values
for bar in bars2:
    height = bar.get_height()
    ax2.annotate(f'{height:.2f}s',
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

# Add combined legend
fig.legend(loc="upper left", bbox_to_anchor=(0.08, 0.80))

# Title and adjust layout
plt.title('Adversarial Method Comparison: Accuracy vs. Efficiency', fontsize=14, pad=20)
plt.tight_layout()

plt.savefig("adversarial_method_comparison.svg", format="svg", bbox_inches = "tight")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
datasets = ['IMDB', 'MNIST', 'ADAS']
xai_methods = ['\\xaiSG', '\\xaiIG', '\\xaiLime', 'SmoothGrad', 'GradCam++', 'IntegratedGrad']

# Overhead percentages (order: IMDB, MNIST, ADAS)
overhead = {
    '\\xaiSG': [35.2, 87, 19],
    '\\xaiIG': [146.58, 52, 16],
    '\\xaiLime': [23663, None, None],  # MNIST/ADAS not applicable
    'SmoothGrad': [None, 87, 19],      # IMDB covered by \xaiSG
    'GradCam++': [None, 79, 16],
    'IntegratedGrad': [None, 52, 16]
}

# Plot setup
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10), gridspec_kw={'height_ratios': [1, 3]})

# --- Main Plot (Log Scale for IMDB) ---
colors = ['#4e79a7', '#f28e2b', '#e15759', '#76b7b2', '#59a14f', '#edc948']
x = np.arange(len(datasets))  # IMDB(0), MNIST(1), ADAS(2)

# Plot bars for each method
for idx, method in enumerate(xai_methods):
    values = []
    for d in range(3):
        val = overhead[method][d]
        values.append(val if val is not None else 0)

    # Filter valid data points
    mask = np.array([val is not None for val in overhead[method]])
    ax1.bar(x[mask] + idx*0.15, np.array(values)[mask], width=0.15,
            color=colors[idx], label=method)

ax1.set_yscale('log')  # Handle \xaiLime's extreme value
ax1.set_ylabel('Time Overhead (%) - Log Scale', fontsize=10)
ax1.set_xticks(x + 0.45)
ax1.set_xticklabels(datasets)
ax1.legend(loc='upper right', ncol=3)
ax1.set_title('XAI Computational Overhead Comparison (Log Scale for IMDB)', pad=20)

# --- Inset Plot (Linear Scale without \xaiLime) ---
ax2 = plt.subplot(2, 1, 2)
filtered_methods = [m for m in xai_methods if m != '\\xaiLime']

for idx, method in enumerate(filtered_methods):
    values = [overhead[method][d] or 0 for d in range(3)]
    ax2.bar(x + idx*0.2, values, width=0.2, color=colors[idx], label=method)

ax2.set_ylabel('Time Overhead (%) - Linear Scale', fontsize=10)
ax2.set_ylim(0, 160)
ax2.set_xticks(x + 0.5)
ax2.set_xticklabels(datasets)
ax2.axhline(y=100, color='grey', linestyle='--', linewidth=0.8)  # 100% reference line

# Annotations
ax1.text(0.02, 0.85, '*\\xaiLime on IMDB: 23,663%', transform=ax1.transAxes, color='red')
ax2.text(1.5, 140, '\\deepjanus Baseline', ha='center', color='purple')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches

# Data with unified method names
imdb_methods = ['SmoothGrad', 'Integrated Gradients', 'LIME']
imdb_overhead = [35.2, 146.58, 23663]  # xaiSG → SmoothGrad，xaiIG → Integrated Gradients

mnist_methods = ['SmoothGrad', 'GradCam++', 'Integrated Gradients']
mnist_overhead = [87, 79, 52]

adas_methods = ['SmoothGrad', 'GradCam++', 'Integrated Gradients']
adas_overhead = [19, 16, 16]

# Color mapping for unified legend
method_colors = {
    'SmoothGrad': '#1f77b4',
    'Integrated Gradients': '#ff7f0e',
    'GradCam++': '#2ca02c',
    'LIME': '#d62728'
}

# Create subplots
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4.5))

# ---------------------------
# IMDB Dataset
# ---------------------------
# Generate colors based on method names
imdb_palette = [method_colors[method] for method in imdb_methods]
sns.barplot(x=imdb_methods, y=imdb_overhead, ax=ax1, palette=imdb_palette)
ax1.set_title('IMDB Dataset', fontsize=20, pad=20)
ax1.set_ylabel('Overhead (%)', fontsize=20)
ax1.set_yscale('log')  # Keep log scale for xaiLime
ax1.set_xticklabels([])  # Remove x-axis labels
ax1.tick_params(axis='y', labelsize=16)  # Set y-axis tick font size to 14
# Add value labels
for i, value in enumerate(imdb_overhead):
    ax1.text(
        i, value,
        f'{value:.2f}%' if value > 1000 else f'{value}%',  # 针对LIME的特殊格式化
        ha='center',
        va='bottom',
        fontsize=15,
        color='black'
    )

# ---------------------------
# MNIST Dataset
# ---------------------------
mnist_palette = [method_colors[method] for method in mnist_methods]
sns.barplot(x=mnist_methods, y=mnist_overhead, ax=ax2, palette=mnist_palette)
ax2.set_title('MNIST Dataset', fontsize=20, pad=20)
ax2.set_ylabel('')
ax2.set_xticklabels([])  # Remove x-axis labels
ax2.tick_params(axis='y', labelsize=16)  # Set y-axis tick font size to 14
# Add value labels
for i, value in enumerate(mnist_overhead):
    ax2.text(
        i, value,
        f'{value}%',
        ha='center',
        va='bottom',
        fontsize=18,
        color='black'
    )

# ---------------------------
# ADAS Dataset
# ---------------------------
adas_palette = [method_colors[method] for method in adas_methods]
sns.barplot(x=adas_methods, y=adas_overhead, ax=ax3, palette=adas_palette)
ax3.set_title('ADAS Dataset', fontsize=20, pad=20)
ax3.set_ylabel('')
ax3.set_xticklabels([])  # Remove x-axis labels
ax3.tick_params(axis='y', labelsize=16)  # Set y-axis tick font size to 14
# Add value labels
for i, value in enumerate(adas_overhead):
    ax3.text(
        i, value,
        f'{value}%',
        ha='center',
        va='bottom',
        fontsize=18,
        color='black'
    )

# ---------------------------
# Global Legend
# ---------------------------
legend_handles = [
    mpatches.Patch(color=method_colors['SmoothGrad'], label='SmoothGrad'),
    mpatches.Patch(color=method_colors['Integrated Gradients'], label='Integrated Gradients'),
    mpatches.Patch(color=method_colors['GradCam++'], label='GradCam++'),
    mpatches.Patch(color=method_colors['LIME'], label='LIME')
]

# Add legend below subplots
fig.legend(
    handles=legend_handles,
    loc='center right',
    ncol=1,
    bbox_to_anchor=(1.25, 0.5),  # 调整位置
    frameon=False,
    fontsize=18
)

# Adjust layout
plt.tight_layout()
plt.subplots_adjust(bottom=0.15)  # 为图例腾出空间
sns.despine()  # 移除多余边框
plt.savefig("xai_overhead.pdf", format="pdf", bbox_inches = "tight")
plt.show()
```
